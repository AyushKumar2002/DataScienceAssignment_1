```python
import random
import matplotlib.pyplot as mplot
import numpy as npy
import pandas as pd
import math
import collections
```

```python
#q1
#Assigning values between 60 and 70 and randomly choosing 100 values within this range
datas = [i for i in range(60, 70)]
[datas.append(random.choice(datas)) for i in range(1000)]
print(datas)
```

```python
#Creating series to store values
dt = pd.Series(datas)
dt
```

```python
#q2
#Calculating mean and standard deviation of the datas
mean = dt.mean()
print(f'Mean = {mean}')

std_deviation = dt.std()
print(f'Standard Deviation = {std_deviation}')
```

```python
#q3
#Plotting histogram of the distribution
mplot.hist(datas, edgecolor='black')
mplot.title('Historgram of 1000 Random values between 60 and 70')
mplot.xlabel('Values')
mplot.ylabel('Frequency')
mplot.show()
```

```python
#q4 and q5
sample = []
smean = []

for i in range(30):
    for j in range(30):
        val = random.choice(datas)
        sample.append(val)
    temp = npy.mean(sample)
    smean.append(temp)
print(smean)
```
